```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
import datetime
%matplotlib inline
```

```python
# change to different json files can get the information
df = pd.read_json('2021_01_24.json')
df.head()
```

```python
# Find the range in the GPS_LONGITUDE column

longitude = df['GPS_LONGITUDE']
longitude_set = set()

for i in range(len(longitude)):
    try:
        longitude_set.add(float(longitude[i]))
    except:
        continue

# print(len(longitude_set))
longitude_list = sorted(longitude_set)

print(longitude_list[0]) # The minimum one
print(longitude_list[-1]) # The maximum one
```

```python
# Find the range in the GPS_LATITUDE column

latitude = df['GPS_LATITUDE']
latitude_set = set()

for i in range(len(latitude)):
    try:
        latitude_set.add(float(latitude[i]))
    except:
        continue

# print(len(longitude_set))
latitude_list = sorted(latitude_set)

print(latitude_list[0])  # The minimum one
print(latitude_list[-1]) # The maximum one
```

```python
# Find the range in the EVENT_NO_TRIP column

event_no_trip = df['EVENT_NO_TRIP']
event_set = set()

for i in range(len(event_no_trip)):
    event_set.add(event_no_trip[i])

event_list = sorted(event_set)
print(event_list)
```

```python
# Find the range in the GPS_SATELLITES column

gps_satellites = df['GPS_SATELLITES']
gps_satellites_set = set()

for i in range(len(gps_satellites)):
    gps_satellites_set.add(gps_satellites[i])

gps_satellites_list = sorted(gps_satellites_set)
print(gps_satellites_list)
```

```python
# Find the range in the GPS_HDOP 

gps_hdop = df['GPS_HDOP']
gps_hdop_set = set()

for i in range(len(gps_hdop)):
    gps_hdop_set.add(gps_hdop[i])

gps_hdop_list = sorted(gps_hdop_set)
print(gps_hdop_list)
```

```python
# Find the range in the SCHEDULE_DEVIATION 

schedule_deviation = df['SCHEDULE_DEVIATION']
schedule_deviation_set = set()

for i in range(len(schedule_deviation)):
    if schedule_deviation[i]: 
        schedule_deviation_set.add(schedule_deviation[i])

# print(schedule_deviation_set)

schedule_deviation_list = sorted(schedule_deviation_set)
print(schedule_deviation_list)
```

```python
# Find the range of VELOCITY

velocity = df['VELOCITY']
velocity_set = set()

for i in range(len(velocity)):
    if velocity[i]:
        velocity_set.add(velocity[i])

velocity_list = sorted(velocity_set)
print(velocity_list)
```

```python
# Find the range in DIRECTION

direction = df['DIRECTION']
direction_set = set()

for i in range(len(direction)):
    if direction[i]:
        direction_set.add(direction[i])

direction_list = sorted(direction_set)
print(direction_list)
```

```python
# Find the range in METERS

meters = df['METERS']
meters_set = set()

for i in range(len(meters)):
    if meters[i]:
        meters_set.add(meters[i])

meters_list = sorted(meters_set)
print(min(meters_list))
print(max(meters_list))
```

```python
# Find the VEHICLE_ID

vehicle_id = df['VEHICLE_ID']
vehicle_set = set()

for i in range(len(vehicle_id)):
    if vehicle_id[i]:
        vehicle_set.add(vehicle_id[i])

vehicle_list = sorted(vehicle_set)
print(vehicle_list)
print(len(vehicle_list))
```
